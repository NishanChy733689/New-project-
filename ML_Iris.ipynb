## Loading the dataset

```python
df=sns.load_dataset('iris')
```

## Mining the data

## Making the regression Model

```python
model=LinearRegression()
```

```python
y=df.species
```

```python
x=df.drop('species',axis=1)
```

```python
y.replace({'virginica':1,'setosa':2,'versicolor':3},inplace=True)
```

```python
trainx, testx, trainy, testy = train_test_split(x, y, test_size=0.2, )
```

```python
trainx
```

```python
trainy
```

```python
model.fit(trainx,trainy)
```

```python
predictions=list(model.predict(testx))
```

```python
prediction_df=pd.DataFrame()#({'prediction':[]})
```

```python
prediction_df['prediction']=predictions
```

```python
pure=[]
for i in prediction_df.prediction:
  if i >= 1 and i<=2:
    pure.append('virginica')
  elif i>=2 and i <=3:
    pure.append('setosa')
  elif i == 3:
    pure.append('versicolor')
```

```python
pure
```

```python
pure=[]
for i in testy:
  if i >= 1 and i<=2:
    pure.append('virginica')
  elif i>=2 and i <=3:
    pure.append('setosa')
  elif i==3:
    pure.append('versicolor')
```

```python
pure
```
